```python
def return_missing_values(data_frame):
    missing_values = data_frame.isnull().sum()/len(data_frame)
    missing_values = missing_values[missing_values>0]
    missing_values.sort_values(inplace=True)
    return missing_values

def plot_missing_values(data_frame):
    missing_values = return_missing_values(data_frame)
    missing_values = missing_values.to_frame()
    missing_values.columns = ['numbers_of_like']
    missing_values.index.names = ['id']
    missing_values['id'] = missing_values.index
    sns.set(style='darkgrid')
    sns.barplot(x='id', y='numbers_of_like', data=missing_values)
    plt.xticks(rotation=90)
    plt.show()
```

```python
def return_unique_values(data_frame):
    unique_dataframe = pd.DataFrame()
    unique_dataframe['Features'] = data_frame.columns
    uniques = []
    for col in data_frame.columns:
        u = data_frame[col].nunique()
        uniques.append(u)
    unique_dataframe['Uniques'] = uniques
    return unique_dataframe
```

```python
def plot_frequency_charts(df, feature, title, pallete):
    freq_df = pd.DataFrame()
    freq_df[feature] = df[feature]
    
    f, ax = plt.subplots(1,1, figsize=(16,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette=pallete)
    g.set_title("Number and percentage of {}".format(title))

    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 

    plt.title('top {}  tweeting '.format(feature))
    plt.ylabel('Frequency', fontsize=12)
    plt.xlabel(title, fontsize=12)
    plt.xticks(rotation=90)
    plt.show()
```

```python
from string import punctuation
from nltk.corpus import stopwords
print(stopwords.words('english')[10:15])

def punctuation_stopwords_removal(sms):
    # filters charecter-by-charecter : ['h', 'e', 'e', 'l', 'o', 'o', ' ', 'm', 'y', ' ', 'n', 'a', 'm', 'e', ' ', 'i', 's', ' ', 'p', 'u', 'r', 'v', 'a']
    remove_punctuation = [ch for ch in sms if ch not in punctuation]
    # convert them back to sentences and split into words
    remove_punctuation = "".join(remove_punctuation).split()
    filtered_sms = [word.lower() for word in remove_punctuation if word.lower() not in stopwords.words('english')]
    return filtered_sms
```

```python
from collections import Counter

def draw_bar_graph_for_text_visualization(df, location):
    tweets_from_loc = df.loc[df.author==location]
    tweets_from_loc.loc[:, 'content'] = tweets_from_loc['content'].apply(punctuation_stopwords_removal)
    loc_tweets_curated = tweets_from_loc['content'].tolist()
    loc_tweet_list = []
    for sublist in loc_tweets_curated:
        for word in sublist:
            loc_tweet_list.append(word)
    loc_tweet_count = Counter(loc_tweet_list)
    loc_top_30_words = pd.DataFrame(loc_tweet_count.most_common(50), columns=['word', 'Counter'])
    fig, ax = plt.subplots(figsize=(16, 6))
    sns.barplot(x='word', y='Counter', 
                data=loc_top_30_words, ax=ax)
    plt.title("Top 50 Most Prevelant Words in {}".format(location))
    plt.xticks(rotation='vertical');
```

```python
from wordcloud import WordCloud, STOPWORDS
def draw_word_cloud(df, location, title):
    loc_df = df.loc[df.author==location]
    loc_df.loc[:, 'content'] = loc_df['content'].apply(punctuation_stopwords_removal)
    word_cloud = WordCloud(
                    background_color='white',
                    stopwords=set(STOPWORDS),
                    max_words=50,
                    max_font_size=40,
                    scale=5,
                    random_state=1).generate(str(loc_df['content']))
    fig = plt.figure(1, figsize=(10,10))
    plt.axis('off')
    fig.suptitle(title, fontsize=20)
    fig.subplots_adjust(top=2.3)
    plt.imshow(word_cloud)
    plt.show()
```

```python
from collections import Counter

def draw_test_bar(df, location):
    df = df.dropna()

    tweets_from_loc = df.loc[df.author==location]
    loc_tweets_curated = tweets_from_loc['content'].tolist()

    tweets_from_loc.loc[:, 'content'] = tweets_from_loc['content'].apply(punctuation_stopwords_removal)
    loc_shares_curated = tweets_from_loc['number_of_shares'].tolist()
    loc_likes_curated = tweets_from_loc['number_of_likes'].tolist()


    loc_tweet_likes_and_shares = []
    for current in range(0,len(loc_tweets_curated)):
        loc_tweet_likes_and_shares.append((loc_likes_curated[current],loc_shares_curated[current],loc_tweets_curated[current]))


    loc_tweets = pd.DataFrame(loc_tweet_likes_and_shares, columns=['likes', 'shares','content'],index=loc_tweets_curated)
      
    loc_tweets.likes = pd.to_numeric(loc_tweets.likes)
    loc_tweets.shares = pd.to_numeric(loc_tweets.shares)

    loc_tweets= loc_tweets.head(50)


    fig, ax = plt.subplots(figsize=(16, 6))
    sns.barplot(x='content', y='likes', 
                data=loc_tweets, ax=ax)
    plt.title("{} 50 content Likes count".format(location))
    plt.xticks(rotation='vertical');


    fig, ax = plt.subplots(figsize=(16, 6))
    sns.barplot(x='content', y='shares', 
                data=loc_tweets, ax=ax)
    plt.title("{} 50 content Shares count".format(location))
    plt.xticks(rotation='vertical');
```

```python
def Manipulate(x,num_like,num_shares):
  hash=x.count('#')+x.count('@')
  numtweet=len(x)
  num_shares=num_shares
  num_like=num_like
  
  return [num_like,num_shares,hash,numtweet]
```
